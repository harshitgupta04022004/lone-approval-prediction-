```python
def remove_outliers(df, features):
    for feature in features:
        Q1 = df[feature].quantile(0.25)
        Q3 = df[feature].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # Filter data for each feature
        df = df[(df[feature] >= lower_bound) & (df[feature] <= upper_bound)]
    return df

# Apply the function to remove outliers from all specified features
features_to_check = ['person_age', 'person_income', 'person_emp_length', 'loan_amnt', 'loan_int_rate', 'loan_percent_income']
x_check = remove_outliers(x_check, features_to_check)
X['loan_status'] = y
X = remove_outliers(X, features_to_check)
X ,y= X.iloc[:,:-1], X.iloc[:,-1] 
```

```python
for i in [0.000001, 0.00001, 0.002, 0.0003]:
    rig_log = RidgeClassifier(alpha=i,random_state=42)
    rig_log.fit(x_train,y_train)
    pre_rid = rig_log.predict(x_cv)
    pre_train_rid = rig_log.predict(x_train)
    print(i,"step--------------------->")
    print(accuracy_score(pre_train_rid,y_train))
    print(accuracy_score(pre_rid,y_cv))
    print(f1_score(pre_rid,y_cv))
    print(precision_score(pre_rid,y_cv))
    print(recall_score(pre_rid,y_cv))
    print(confusion_matrix(pre_rid,y_cv))
```

```python
best_rforest_n_e = 0
rfor_aqu = 0
for i in range(50,200,50):
    rforest_n_e = RandomForestClassifier(random_state=42,n_estimators=i,)
    rforest_n_e.fit(x_train,y_train)
    pre_rforest_n_e = rforest_n_e.predict(x_cv)
    pre_train_rforest_n_e = rforest_n_e.predict(x_train)
    print(i,"step--------------------->")
    print(accuracy_score(pre_train_rforest_n_e,y_train))
    print(accuracy_score(pre_rforest_n_e,y_cv))
    print(f1_score(pre_rforest_n_e,y_cv))
    print(precision_score(pre_rforest_n_e,y_cv))
    print(recall_score(pre_rforest_n_e,y_cv))
    print(confusion_matrix(pre_rforest_n_e,y_cv))
    aqu = accuracy_score(pre_rforest_n_e,y_cv)
    if rfor_aqu < aqu:
        rfor_aqu = aqu
        best_rforest_n_e = i
    
```
